```python
q = [f'value: {v[0]} counts: {v[1]}' for v in zip(np.unique(df['feature_0'].values),  np.unique(df['feature_0'].values, return_counts=True)[1])]
print(*q, sep='\n')
```

```python
for col in df.columns.values.tolist()[2:]:
    v, c = np.unique(df[col].values, return_counts=True)
    print(f"{col} min: {v.min()} max: {v.max()}")
    print(f'{col} has {len(v)} different values')
    if len(np.unique(df[col].values)) <= 20:
        l = [f'value: {el[0]} counts: {el[1]}' for el in zip(v, c)]
        print(*l, sep='\n')
    plt.bar(v, c)
    plt.xlabel(col +'_values')
    plt.show()
```
